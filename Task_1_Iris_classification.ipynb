```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report,roc_auc_score,roc_curve,auc
```

```python
iris = pd.read_csv('/content/Iris.csv')
print(iris.head())
```

```python
print(iris['Species'].value_counts())
```

```python
sns.countplot(x='Species',data=iris, palette=['red', 'green', 'blue'])
plt.title('countplot of the Species')
plt.show()
```

```python
#Visualization
sns.scatterplot(
    data= iris,
    x="SepalLengthCm", y="SepalWidthCm", hue="Species")
plt.title('Sepal Width vs Sepal Length')
plt.show()
```

```python
#Visualization
sns.scatterplot(
    data= iris,
    x="PetalLengthCm", y="PetalWidthCm", hue="Species")
plt.title('Petal Width vs Petal Length')
plt.show()
```

```python
#Using heatmap to visualize the correlation between the variables
heatmap = iris.drop(['Id','Species'], axis=1)
sns.heatmap(heatmap.corr(),annot=True,cmap='viridis')
plt.title('Iris Heatmap')
plt.show()
```

```python
#To visualize the whole dataset
df = iris.drop('Species', axis=1)
df['species'] = iris['Species']

# Create a parallel coordinates plot
plt.figure(figsize=(10, 6))
pd.plotting.parallel_coordinates(df, 'species', color=('#FF0000', '#00FF00', '#0000FF'))

# Set plot title and labels
plt.title('Parallel Coordinates Plot - Iris Dataset')
plt.xlabel('Features')
plt.ylabel('Feature Values')

# Display the plot
plt.show()
```

```python
#Visualizing the whole dataset using pairplot
sns.pairplot(df, hue='species')
plt.show()
```

```python
X = iris.drop(['Species', 'Id'], axis=1)
y = iris['Species']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42)
```

```python
nb_model = GaussianNB()
```

```python
nb_model.fit(X_train,y_train)
```

```python
y_pred = nb_model.predict(X_test)
```

```python
y_prob = nb_model.predict_proba(X_test)[:,1]
print(classification_report(y_test, y_pred))


ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred)).plot()


nb_model_score = accuracy_score(y_test, y_pred)
print('Accuracy:', nb_model_score)
```

```python
#Testing my prediction
prediction = nb_model.predict(X_test)
print("Prediction of Species: {}".format(prediction))
```
